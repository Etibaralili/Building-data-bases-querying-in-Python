The pairwise deletion was used here to remove columns that have empty spaces and not of interest in analysis at task 3. If we would to keep the variables, we would use listwise deletion.

```python
#Replacing % sign
df1['AVERAGE_STUDENT_ATTENDANCE'] = df1['AVERAGE_STUDENT_ATTENDANCE'].str.replace('%','')
df1['Average Teacher Attendance'] = df1['Average Teacher Attendance'].str.replace('%','')
df1['Individualized Education Program Compliance Rate'] = df1['Individualized Education Program Compliance Rate'].str.replace('%','')
```

```python
pd.set_option('display.max_columns',60)# column limit
pd.set_option('display.max_rows',500)# row limit
```

```python
# Selecting important variables
df1 = df1[['School ID',
    'NAME_OF_SCHOOL',
    'COMMUNITY_AREA_NUMBER',                             
    'AVERAGE_STUDENT_ATTENDANCE',
    'Rate of Misconducts (per 100 students)',
    'Average Teacher Attendance',
    'Individualized Education Program Compliance Rate',
    'COLLEGE_ENROLLMENT',
    'Instruction Score',
    'Adequate Yearly Progress Made?',
    'HEALTHY_SCHOOL_CERTIFIED',
    'Environment Score',
    'SAFETY_SCORE',]]
```

```python
# Converting columns to a numeric datatype 
df1['AVERAGE_STUDENT_ATTENDANCE'] = pd.to_numeric(df1['AVERAGE_STUDENT_ATTENDANCE'])
df1['Individualized Education Program Compliance Rate'] = pd.to_numeric(df1['Individualized Education Program Compliance Rate'])
df1['Average Teacher Attendance'] = pd.to_numeric(df1['Average Teacher Attendance'])
# Converting categorical data to a numeric datatype
df1['Adequate Yearly Progress Made?'] = df1['Adequate Yearly Progress Made?'].replace({'Yes':1,'No':0})
df1['HEALTHY_SCHOOL_CERTIFIED'] = df1['HEALTHY_SCHOOL_CERTIFIED'].replace({'Yes':1,'No':0})


df1.head()
```

```python
df1_sql = sqlite3.connect("df1")
try:
  df1_sql.execute('''
      CREATE TABLE Public_School
      (School_ID                                         INTEGER  PRIMARY KEY,
       NAME_OF_SCHOOL                                    TEXT     NOT NULL,
       COMMUNITY_AREA_NUMBER                             INTEGER  DEFAULT 0,
       AVERAGE_STUDENT_ATTENDANCE                        FLOAT    DEFAULT 0,
       Rate_of_Misconducts_per_100_students              INTEGER  DEFAULT 0,
       Average_Teacher_Attendance                        FLOAT    DEFAULT 0,
       Individualized_Education_Program_Compliance_Rate  FLOAT    DEFAULT 0, 
       COLLEGE_ENROLLMENT                                INTEGER  DEFAULT 0,
       Instruction_Score                                 FLOAT    DEFAULT 0,
       Adequate_Yearly_Progress_Made                    INTEGER  DEFAULT 0,
       HEALTHY_SCHOOL_CERTIFIED                          INTEGER  DEFAULT 0,
       Environment_Score                                 FLOAT    DEFAULT 0,
       SAFETY_SCORE                                      FLOAT    DEFAULT 0,
       FOREIGN KEY (COMMUNITY_AREA_NUMBER) REFERENCES Census)
        ;''')
  print("Table created successfully")
except Exception as e:
  print(str(e))
  print('Table creation failed!!!')
finally:
  df1_sql.close()
```

```python
plt.scatter(df_all['Average Teacher Attendance'], df_all['Individualized Education Program Compliance Rate'])
z = np.polyfit(df_all['Average Teacher Attendance'], df_all['Individualized Education Program Compliance Rate'], 1)
p = np.poly1d(z)
plt.plot(df_all['Average Teacher Attendance'],p(df_all['Average Teacher Attendance']),"r--")

plt.show()

from scipy.stats.stats import pearsonr
corr, _ = pearsonr(df_all['Average Teacher Attendance'], df_all['Individualized Education Program Compliance Rate'])
print('Pearsons correlation: %.3f' % corr)
```

The relationship between individualied eduation program compliance rate and teacher attendance was explored here. The above plot and pearson test indicate the the relationship between the variables are positive (if the individualized eduaiton program compliance rate increates the average teacher attendance increases) and weak (between 0.1 and 0.3).
